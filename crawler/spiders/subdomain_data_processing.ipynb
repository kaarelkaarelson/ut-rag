```python
import json

with open('../../subdomains_WhoisXMLAPI.json', 'r') as file:
    data = json.load(file)

records = data['result']['records']

domains = [record['domain'] for record in records ]

for d in domains:
    print(d)

with open('../../subdomains_clean.txt', 'w') as file:
    for d in domains:
        file.write(d + '\n')

```
